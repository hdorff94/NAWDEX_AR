## Get east and west domain data

```python
import cdsapi
import os
#import numpy as np
#import pandas as pd
#import xarray as xr
c = cdsapi.Client()

east_file='east_20150314_09.nc'
west_file='west_20150314_09.nc'
```

### east

```python
if not os.path.exists(east_file):
    c.retrieve(
        'reanalysis-carra-single-levels',
        {
            'format': 'netcdf',
            'variable': [
                '10m_u_component_of_wind', '10m_v_component_of_wind', '10m_wind_direction',
                '10m_wind_speed', 'total_precipitation',
            ],
            'product_type': 'forecast',
            'time': '09:00',
            'leadtime_hour': '1',
            'year': '2015',
            'month': '03',
            'day': '14',
            'domain': 'east_domain',
            'level_type': 'surface_or_atmosphere',
        },
        east_file)
else:
    pass
```
